```python
# %load /usr/local/bin/csongs
#!/usr/bin/python3

import os
import csv
import json
from collections import defaultdict

path = r'/home/karen/Documentos/javier/git/linux_conf/songs/'


def get_csv(file_name):
    return list(csv.reader(open(os.path.join(path, file_name), 'r', encoding='utf-8-sig')))


list_files = ['romanticas.txt', 'chidas.txt', 'rock.txt']


def fun(lista):
    return [n for n in lista]


rom, ch, ro = fun(list_files)
bands = {}


def make(*args):
    for p in args:
        for n in get_csv(p):
            try:
                band, *_ = n[0].split(' - ')
                song = [' '.join(_)] if len(_) > 1 else _
            except:
                pass
            else:
                bands.setdefault(band, {
                    'total': 0,
                    'dir': p.replace('.txt', ''),
                    'songs': []
                })
                bands[band]['total'] += 1
                bands[band]['songs'] += song

make(rom, ch, ro)

# for k, v in bands.items():
#     print(f'Band: {k}, Total: {v["total"]}')
#     for i in v['songs']:
#         print(i)

_totals = {}
_totals1 = defaultdict(int)

for k,v in bands.items():
    _totals.setdefault(v['dir'], 0)
    _totals[v['dir']] += v['total']
# print(_totals)

# for k, v in bands.items():
#     _totals1[v['dir']] += v['total']
# print(_totals1)

# no usar .items() y acceder directo al diccinario
for s in bands:
    _totals1[bands[s]['dir']] += bands[s]['total']
print(_totals1)

#with open('songs.json', 'w', encoding='utf-8-sig') as f:
#    json.dump(bands, f, ensure_ascii=False, indent=2)
```
